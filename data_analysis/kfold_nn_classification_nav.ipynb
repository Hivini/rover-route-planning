```python
import numpy as np

from sklearn import datasets
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils

# Evaluate model
kf = KFold(n_splits=5, shuffle = True)

x = np.load('nav_data.npy')
y = np.load('nav_data_labels.npy')
y = y - 1

n_clases = 4
n_features = len(x[0])
```

```python
from keras.callbacks import EarlyStopping
acc = 0
recall = np.zeros(4)
for train_index, test_index in kf.split(x):

    # Training phase
    x_train = x[train_index, :]
    y_train = y[train_index]
    y_train = np_utils.to_categorical(y_train)

    clf = Sequential()
    clf.add(Dense(128, input_dim=17, activation='relu'))
    clf.add(Dense(64, activation='relu'))
    clf.add(Dense(16, activation='relu'))
    clf.add(Dense(4, activation='softmax'))

    es = EarlyStopping(monitor='loss', mode='min', verbose=1, patience=20)
    clf.compile(loss='categorical_crossentropy', optimizer='adam')
    clf.fit(x_train, y_train, epochs=100, batch_size=4, callbacks=[es])
    print()
    # Test phase
    x_test = x[test_index, :]
    y_test = y[test_index]
    y_pred = np.argmax(clf.predict(x_test), axis=-1)

    cm = confusion_matrix(y_test, y_pred)

    acc += (cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3])/len(y_test)    

    recall[0] += cm[0,0]/(cm[0,0] + cm[0,1] + cm[0,2] + cm[0, 3])
    recall[1] += cm[1,1]/(cm[1,0] + cm[1,1] + cm[1,2] + cm[1, 3])
    recall[2] += cm[2,2]/(cm[2,0] + cm[2,1] + cm[2,2] + cm[2, 3])
    recall[3] += cm[3,3]/(cm[3,0] + cm[3,1] + cm[3,2] + cm[3, 3])


acc = acc/5
print('ACC = ', acc)

recall = recall/5
print('RECALL = ', recall)
```
